```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.stats import poisson
```

# Week 3

#### Example for simulation binomial distribution - please see the Homework 2 - Example 2

```python
# Generate random values from a binomial distribution
sample = np.random.binomial(n=21, p=0.1, size=10000)

# Create a histogram
plt.hist(sample, color="orange", edgecolor="black")

# Set plot properties
plt.title("number of lates")
plt.xlabel("possible values")
plt.ylabel("probabilities")

# Show the plot
plt.show()
```

```python
sample_mean = np.mean(sample)
sample_std = np.std(sample)

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
```

```python
# Calculate the probability of exactly 2 successes
# k=2 (number of successes), n=21 (number of trials), p=0.1 (probability of success)
probability = binom.pmf(k=2, n=21, p=0.1)

print(probability)
```

```python
# Generate 200 random values from a Poisson distribution
sample = np.random.poisson(lam=3.42, size=200)

# Define the bins to match the R code's 'breaks'
# The R breaks c(-0.5, 0.5, 1.5, ...) are used for integer-centered bins
bins = np.arange(np.min(sample), np.max(sample) + 2) - 0.5

# Create a histogram with the specified bins
plt.hist(sample, bins=bins, color="orange", edgecolor="black")

# Set the plot properties to match the R code
plt.title("Poisson distribution")
plt.xlabel("possible values")
plt.ylabel("frequencies")

# Show the plot
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
sample_mean = np.mean(sample)
sample_std = np.std(sample)

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
```

```python
relative_frequency = np.sum(sample == 3) / len(sample)
print(relative_frequency)
```

```python
# Calculate the cumulative probability of P(X <= 3) for a Poisson distribution
# with lambda (mu) = 3.42
probability = poisson.cdf(k=3, mu=3.42)

print(probability)
```

```python
# Calculate the probability of exactly 3 events
# k=3 (number of events), mu=3.42 (lambda or mean)
probability = poisson.pmf(k=3, mu=3.42)

print(probability)
```
